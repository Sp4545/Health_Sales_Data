```python
import pandas as pd

# Load datasets with encoding handling
pbj_df = pd.read_csv("pbj.csv", encoding="ISO-8859-1")
provider_df = pd.read_csv("provider.csv", encoding="ISO-8859-1")


# Display initial dataset shapes
print("PBJ Staffing Data Shape:", pbj_df.shape)
print("Provider Data Shape:", provider_df.shape)
```

```python
datasets = [pbj_df, provider_df]

# Function to standardize column names
def standardize_column_names(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
    return df

# Apply column name standardization
for name in datasets:
    name = standardize_column_names(name)
```

```python
# Function to handle missing values
def handle_missing_values(df):
    for col in df.columns:
        if df[col].isnull().sum() > 0:  # If missing values exist
            if df[col].dtype == 'object':  # Categorical column
                df[col].fillna(df[col].mode()[0], inplace=True)
            else:  # Numerical column
                df[col].fillna(df[col].median(), inplace=True)
    return df
```

```python
# Function to convert data types
def convert_data_types(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].astype(str)  # Ensure categorical columns are strings
        elif 'date' in col.lower():
            df[col] = pd.to_datetime(df[col], errors='coerce')  # Convert date columns
    return df

# Apply data type conversion
for name in datasets:
    name = convert_data_types(name)
    print(name.dtypes)

# Check for missing values after handling
for df in datasets:
    # print(f"Missing values in {name}:")
    print(df.dtypes, "\n")
```

```python
# Function to remove duplicate rows
def remove_duplicates(df):
    return df.drop_duplicates()

dataset_new = []
# Apply duplicate removal correctly
for name in datasets:
    name = handle_missing_values(name)
    dataset_new.append(remove_duplicates(name)) # Update the dictionary correctly
    # print(name.duplicated().sum())

# Check the number of duplicate values after removal
for df in dataset_new:
    duplicate_count = df.duplicated().sum()
    print(duplicate_count)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Aggregate contractor reliance by state
statewise_contractor_usage = merged_df.groupby("state")[["hrs_rn_ctr", "hrs_lpn_ctr", "hrs_cna_ctr"]].mean()

# Create a heatmap
plt.figure(figsize=(15,15))
sns.heatmap(statewise_contractor_usage, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Contractor Staffing Reliance by State")
plt.show()
```

```python
from matplotlib.table import Table

# Group by 'provname' and sum the contractor CNA hours, then get the top 10 distinct facilities
top_facilities = merged_df.groupby("provname")["hrs_cna_ctr"].sum().nlargest(10).reset_index()

# Extract short names (first and last names)
def extract_short_name(full_name):
    parts = full_name.split(', ')
    short_names = []
    for part in parts:
        words = part.split()
        if len(words) > 1:
            short_names.append(' '.join([words[0], words[-1]]))
        else:
            short_names.append(words[0])
    return ', '.join(short_names)

top_facilities['short_name'] = top_facilities['provname'].apply(extract_short_name)

# Plotting vertically
plt.figure(figsize=(16, 7))  # Increased overall width for more table space
gs = plt.GridSpec(1, 2, width_ratios=[1.5, 1])  # Adjusted width ratios

# Bar plot
ax1 = plt.subplot(gs[0])
sns.barplot(x=top_facilities["short_name"], y=top_facilities["hrs_cna_ctr"], palette="viridis", ax=ax1)
ax1.set_title("Top 10 Facilities with Highest CNA Contractor Reliance")
ax1.set_ylabel("CNA Contractor Hours")
ax1.set_xlabel("Facility Name (Short)")
plt.xticks(rotation=45, ha='right')

# Pivot table (as matplotlib.table.Table)
ax2 = plt.subplot(gs[1])
ax2.axis('off')

table = Table(ax2, loc='center')

# Add headers
cell = table.add_cell(0, 0, width=0.6, height=0.1, text="Facility Name", loc='center', facecolor='lightgray') # increased header width
cell.get_text().set_fontsize(12)
cell = table.add_cell(0, 1, width=0.4, height=0.1, text="CNA Hours", loc='center', facecolor='lightgray') # decreased cna hour width
cell.get_text().set_fontsize(12)

# Add data rows
for i, row in top_facilities.iterrows():
    # Split facility names into two lines if needed
    full_name = row['provname']
    if len(full_name.split()) > 3:  # Adjust this threshold as needed
        words = full_name.split()
        mid_point = len(words) // 2
        line1 = ' '.join(words[:mid_point])
        line2 = ' '.join(words[mid_point:])
        display_name = f"{line1}\n{line2}"
    else:
        display_name = full_name

    cell = table.add_cell(i + 1, 0, width=0.6, height=0.1, text=display_name, loc='left') # increased the cell width.
    cell.get_text().set_fontsize(10)
    cell = table.add_cell(i + 1, 1, width=0.4, height=0.1, text=f"{row['hrs_cna_ctr']:.2f}", loc='right') # decreased the cna hour width
    cell.get_text().set_fontsize(10)

# Add table to the axis
ax2.add_table(table)

# Make the table visible
table.auto_set_font_size(False)
table.scale(1, 1.5)  # Adjust scaling as needed

plt.tight_layout()
plt.show()
```

```python
# Convert necessary columns to numeric for analysis
numeric_columns = ["hrs_rn_ctr", "hrs_lpn_ctr", "hrs_cna_ctr"]
merged_df[numeric_columns] = merged_df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Identifying top facilities by total contractor hours (RN, LPN, CNA combined)
merged_df["Total_Contractor_Hours"] = merged_df["hrs_rn_ctr"] + merged_df["hrs_lpn_ctr"] + merged_df["hrs_cna_ctr"]

# Selecting top 10 facilities with the highest contractor hours
top_contractor_facilities = merged_df.groupby("provname")["Total_Contractor_Hours"].sum().nlargest(10).reset_index()

# Plotting
plt.figure(figsize=(12,6))
sns.barplot(x=top_contractor_facilities["Total_Contractor_Hours"], y=top_contractor_facilities["provname"], orient='h', palette="Reds_r")
plt.title("Top 10 Facilities by Contractor Staffing Hours")
plt.xlabel("Total Contractor Hours (RN + LPN + CNA)")
plt.ylabel("Facility Name")
plt.show()
```

```python
# Calculate total staffing hours per resident per day
merged_df['total_staffing_hours_per_resident_per_day'] = (
    merged_df['reported_rn_staffing_hours_per_resident_per_day'] +
    merged_df['reported_lpn_staffing_hours_per_resident_per_day'] +
    merged_df['reported_nurse_aide_staffing_hours_per_resident_per_day']
)

# Pivot table: State vs. Average Staffing Hours
pivot_staffing_state = pd.pivot_table(
    merged_df,
    values=['total_staffing_hours_per_resident_per_day'],
    index='state',
    aggfunc='mean'
)

# Sort and get the top 10 states
top_10_staffing = pivot_staffing_state.sort_values(
    by='total_staffing_hours_per_resident_per_day',
    ascending=True
).tail(10)

# Plotting with Seaborn barplot and sequential color palette
plt.figure(figsize=(8, 5))  # Reduced figure size
sns.barplot(
    y=top_10_staffing.index,
    x=top_10_staffing['total_staffing_hours_per_resident_per_day'],
    palette="Blues"  # Use a sequential palette like "Blues" or "Greens"
)

plt.title('Top 10 States vs. Average Staffing Hours', fontsize=12)  # Reduced title font size
plt.xlabel('Average Total Staffing Hours per Resident per Day', fontsize=10)  # Reduced x-axis label font size
plt.ylabel('State', fontsize=10)  # Reduced y-axis label font size
plt.xticks(fontsize=8) # reduced x-axis tick fontsize
plt.yticks(fontsize=8) # reduced y-axis tick fontsize
plt.tight_layout()
plt.show()
```

```python
# Pivot table: Ownership type vs average turnover
pivot_turnover_ownership = pd.pivot_table(
    merged_df,
    values=['total_nursing_staff_turnover', 'registered_nurse_turnover'],
    index='ownership_type',
    aggfunc='mean'
)

# Plotting
plt.figure(figsize=(12, 7))  # Increased width only

# Side-by-side bars
ax = pivot_turnover_ownership.plot(kind='bar', figsize=(12, 7), width=0.8)

plt.title('Avg. Turnover by Ownership', fontsize=14)
plt.ylabel('Turnover (%)', fontsize=12)
plt.xlabel('Ownership', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.legend(['Nursing Staff', 'RN Turnover'], fontsize=10)

# Add percentage labels with slightly larger font
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 7), textcoords='offset points', fontsize=8)

plt.tight_layout()
plt.show()
```
